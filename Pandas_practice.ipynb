```python
import pandas  as pd
runs = [23,56,78]
pd.Series(runs)
```

```python
# custom index
marks = [45,67,89]
subjects =['math','englis','hindi']
pd.Series(marks,index=subjects)
```

```python
# customer code
marks = [34,56,78]
subjects =['maths','englsh','science']
pd.Series(marks,index=subjects)
```

```python
# setting a name
marks = pd.Series(marks,index = subjects,name = 'Nitish ke marks' )
print(marks)
```

Series from dict


```python
marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}
marks_series = pd.Series(marks,name ='akanksha ke marks')
marks_series
```
